```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('/workspaces/cgeniusIA/dataset_versions/DatasetIA_v1.0.0.csv')
dados.head()
```

```python
dados.columns
```

```python
media_gastos_categoria = dados.groupby('Categoria')['Gastos_Mensais'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(media_gastos_categoria['Categoria'], media_gastos_categoria['Gastos_Mensais'], color='skyblue')
plt.title('Média de Gastos Mensais por Categoria de Produto')
plt.xlabel('Categoria')
plt.ylabel('Gastos Mensais Médios')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('grafico_gastos_categoria.png')
plt.show()
```
